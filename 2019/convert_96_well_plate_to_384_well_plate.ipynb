### Write out a separate function to add these columns to the `tidy_plates` dataframe

Sort the plates by row letter and column number so the reassignment happens in the proper order

```python
tidy_plates = tidy_plates.sort_values(['row_letter', 'column_number'])
tidy_plates.head()
```

Assign the grouped dataframe to its own variable for safekeeping and reducing typing (which causes typos) later

```python
grouped = tidy_plates.groupby(['row_letter', 'column_number'], 
                              group_keys=False)
grouped
```

We can check the number of items in each one to make sure the groups got formed properly

```python
grouped.size()
```

The `grouped` is a `pandas.DataFrameGroupBy` object that we can't "see" the way we would with a regular dataframe but we can apply functions to it

```python
def get_384_column_number(ninety_six_column_number, verbose=False):
    original_index = ninety_six_column_number.index
    
    # This is operated on the whole mini-dataframe column number which 
    # is by definition the same since we made mini-dataframes with groupby()
    # So we just need to extract the first one, aka the 0th index
    ninety_six_column_number = ninety_six_column_number.iloc[0]
    
    column_i = ninety_six_column_number * 2 - 1
    new_column_numbers = list(range(column_i, column_i+2))
    reordered_column_numbers = new_column_numbers + new_column_numbers
    if verbose:
        print(f"converted: {ninety_six_column_number} --> {reordered_column_numbers}")
    return pd.Series(reordered_column_numbers)

grouped['column_number'].apply(get_384_column_number, verbose=True).head()
```

```python

def get_384_row_letter(ninety_six_row_letter, verbose=False):
    # This is operated on the whole mini-dataframe row number which 
    # is by definition the same since we made mini-dataframes with groupby()
    # So we just need to extract the first one, aka the 0th index
    ninety_six_row_letter = ninety_six_row_letter.iloc[0]
    row_i = string.ascii_uppercase.index(ninety_six_row_letter)
    
    new_row_letters = string.ascii_uppercase[row_i*2:(row_i*2+2)]
    reordered_row_letters = [new_row_letters[int(i/2)] for i in range(4)]
    if verbose:
        print(f"converted: {ninety_six_row_letter} --> {reordered_row_letters}")
    return pd.Series(reordered_row_letters)
    
grouped['row_letter'].apply(get_384_row_letter, verbose=True).head()
```

```python
tidy_plates['three_eighty_four_row_letter'] = grouped['row_letter'].apply(get_384_row_letter).values
tidy_plates['three_eighty_four_column_number'] = grouped['column_number'].apply(get_384_column_number).values
tidy_plates.head()
```
