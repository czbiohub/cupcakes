## Histogram Manipulations:
* Sometimes your images come out too dark, with low contrast, or just not optimized for what you want to image for. In these cases you can manipulate your lookup table and/or your histogram by using the built in "clim" function from matplotlib as well as the exposure modules from scikit-image


* **Histogram equalization**: equalizes the histogram across the x-axis (intensity) ultimately allowing the image to have similar amounts of dark spots and light spots rather than concentration intensity in "medium" intensity spots
* **Gamma correction**: adjusts the lookup table to be curved/logarithmic rather than linear to capture more dim or bright structures, depending on your gamma value

## Importing useful libraries

```python
import matplotlib.pyplot as plt
import matplotlib

import numpy as np

import skimage

from skimage import data
from skimage import exposure
```

```python
path = '/Users/liennachan/Documents/'
img = skimage.io.imread(path + 'example_img.tif')
```

## Showing an image

## Plotting a histogram for an image

```python
plt.hist(img.ravel());
```

```python
#increasing the number of bins will refine your histogram results
plt.hist(img.ravel(), bins=100);

#setting titles and x/y labels with matplotlib
plt.title('Histogram')
plt.xlabel('Count')
plt.ylabel('Intensity')
```

## Adjusting contrast and brightness 

```python
plt.imshow(img, clim=(25000, 35000), cmap = 'gray');
```

```python
#clipping the range to display intensity counts for the image above
plt.hist(img.ravel(), bins=100, range=(25000,35000));
```

## Histogram manipulations

```python
#equalizing histogram for an image
img_eq = exposure.equalize_hist(img)
plt.imshow(img_eq, cmap = 'gray');
```

```python
#displaying the histogram after histogram equalization
plt.hist(img_eq.ravel(), bins=100);
```

```python
#gamma greater than 1 will darken the pixels
img_gam2 = exposure.adjust_gamma(img, 2)
plt.imshow(img_gam2, cmap = 'gray')
```

```python
#lower intensities bc darker pixels
plt.hist(img_gam2.ravel(), bins=100);
```

```python
#gamma less than 1 will lighten the image
img_gam02 = exposure.adjust_gamma(img, 0.2)
plt.imshow(img_gam02, cmap = 'gray');
```

```python
#higher intensities bc brighter pixels
plt.hist(img_gam02.ravel(), bins=100);
```
